## Subplot
Mee permite generar más visualizaciones de acuerdo a mis necesidades

Cuando creo un plor 1,2 le indico que es 1 fila y dos columnas, con el index puedo moverme por las posiciones creadas.

````python
plt.subplot(1,2,1) #n filas, n columnas, index para moverme
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.subplot(1,2,2) #en el segundo grafico en index es el 2
plt.hist(y) #definicion del segundo grafico
plt.show
````

````python
plt.subplot(1,2,1) #tres parametros, numero de filas, n de columnas del arreglo grafico y el tercer valor es el index que voy a manejar entre filas y columnas
#es decir ese tercer parámetro indica donde estará este primer gráfico, en este caso indice 1
plt.plot(x,y,'g--') #que quiero en ese grafico que defini
plt.plot(y,x,'b:') #Todo lo que ponga bajo la definición del arreglo grafico saldrá en ese indice (1)
plt.subplot(1,2,2) #en el segundo grafico en index 2
plt.pie(y) #definicion del segundo grafico que saldrá en la posición (2)
plt.show
````

````python
#Modificando disposición de los graficos, ojo con el manejo de los index
plt.subplot(2,1,1)
plt.plot(x,y,'g--')
plt.plot(x,x,'b-')
plt.plot(y,x,'r:')

plt.subplot(2,1,2)
plt.pie(y)
plt.show()
````

## Método Orientado a Objetos

Más constumizable, más a amigable a más diagramas, requiere más código

````python
from matplotlib import axes
#Primero creo el linezo donde vaa a estar mi gráfica
fig = plt.figure()

#Nuestro gráficos serán los "axes", a los cuales le agregamos
axes=fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y,'b')
fig.show()
````

````python
fig, ax1 = plt.subplots()

ax1.set_ylabel('distance (m)')
ax1.set_xlabel('time')
ax1.plot(x,y,'blue')

ax2 = ax1.twinx() #copiamos los valores de x
ax2.set_ylabel('velocity(m/s)')
ax2.set_xlabel('time')
ax2.plot(y,x,'green')

fig.set_size_inches(7,5)
fig.set_dpi(100)
````

````python
#La ventaja viene cuando quiero hacer algo mucho más especifico
fig = plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.9])
# posicion  (horizontal,vertical)
# tamaño    (horizontal,vertical)
axes2=fig.add_axes([0.2,0.6,0.4,0.3]) #para hacer un segundo gráfico

axes.plot(x,y,'b')
axes2.plot(y,x,'g')
axes2.set_facecolor('lightgray') #así puedo modificar el color de fondo de ese axes
#Esta es una ventaja de usar método orientado a objeto, cada objeto es distinto y lo puedo modificar por separado

fig.show()
````

## Subplots

Con "s" al final utiliza las bondades de Orientado a Objetos

````python
x = np.linspace(0,5,11)
y = np.sin(x)
````

````python
fig, axes = plt.subplots() #Defino el lienzo
axes.plot(x,y,'b')
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2) #Defino el lienzo y agrego atributos
#Definí un lienzo con una fila y dos columnas
````

````python
fig, axes = plt.subplots(nrows=1, ncols=2) #Defino el lienzo y agrego atributos
axes[0].plot(x,y,'b') #puedo acceder al axes por su index, pues es una especie de array
axes[1].plot(y,x*100,'g')
````

````python
#Hay otra forma de configurar estos gráficos
#Dado que los axes son definidos como tuplas o array
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2) #los defino al definir el lienzo
ax1.plot(x,y,'b') #luego accedo directamente
ax2.plot(y,x*100,'r')
````

````python
fig, axes = plt.subplots(nrows = 2, ncols=4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,0].set_xlabel('X_label')
axes[0,0].set_ylabel('Y_label')

axes[0,1].plot(x,np.tan(x),'g')
axes[0,2].plot(x,np.tan(y),'b')
axes[0,3].plot(x,np.sin(x)*np.sqrt(np.tan(x))/np.tan(y),'r')

axes[1,0].plot(x,np.cos(y),'orange')
axes[1,1].plot(x,np.sqrt(x),'g')
axes[1,2].plot(x,np.tan(y)**3,'b')
axes[1,3].plot(x,np.sin(x)**-1,'gray')

fig.tight_layout()
plt.show()#Esto es para organizar mejor el lienzo
#Todo esto lo puedo también definir por tuplas como en el ejemplo anterior
````

````python
np.random.seed(19680801)

# example data
mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 50

fig, ax = plt.subplots()

# the histogram of the data
n, bins, patches = ax.hist(x, num_bins, density=True)

# add a 'best fit' line
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
````
